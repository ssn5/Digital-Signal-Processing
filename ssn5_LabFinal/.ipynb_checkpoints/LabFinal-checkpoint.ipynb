## Exercise 1: Building an Edge Detector 2.0

In Lab 2 Exercise 4, we built a simple edge detector by applying a high-pass filter along the rows and columns of an image, then combined the two results to create an image of detected edges. In this exercise, we will build a more sophisticated edge detector by adding onto our original design. This improved detector is known as the [Sobel operator](https://en.wikipedia.org/wiki/Sobel_operator).

The Sobel operator uses the same intuition of finding horizontal and vertical edges separately, then combining these results to form the image. Let $G_x$ be our resulting image from detecting vertical edges (through the x-axis) and $G_y$ be our resulting image from detecting horizontal edges (through the y-axis, then our final result will be given by

$$
G[i,j] = \sqrt{G^2_x[i,j]+G^2_y[i,j]},
$$

where $G[i,j]$ is the pixel value at row $i$, column $j$. We compute $G_x$ and $G_y$ via convolution as follows:

$$
G_x = I * \begin{bmatrix}
1 & 0 & -1\\
\end{bmatrix} * \begin{bmatrix}
1\\
2\\
1\\
\end{bmatrix}
$$

$$
G_y = I * \begin{bmatrix}
1\\
0\\
-1\\
\end{bmatrix} * \begin{bmatrix}
1 & 2 & 1\\
\end{bmatrix},
$$

where $I$ is our original image and $*$ denotes the convolution operator. Note that in the computation of $G_x$ we convolve along the rows with a high-pass filter and convolve along the columns with a low-pass filter, while the computation of $G_y$ reverses this relationship with the same filters.

a. Compute the image $G_x$ by performing the two convolutions along the rows and columns, respectively, using $\textrm{signal.convolve()}$. Plot your resulting image in grayscale. Remember that you should apply the high-pass filter $\begin{bmatrix}1 & 0 & -1\end{bmatrix}$ to each **row** and the low-pass filter $\begin{bmatrix}1 & 2 & 1\end{bmatrix}$ to each **column**. Also, be sure to use the "same" mode when using $\textrm{signal.convolve()}$.

b. Compute the image $G_y$ by performing the two convolutions along the rows and columns, respectively, using $\textrm{signal.convolve()}$. Plot your resulting image in grayscale. Remember that you should apply the high-pass filter $\begin{bmatrix}1 & 0 & -1\end{bmatrix}$ to each **column** and the low-pass filter $\begin{bmatrix}1 & 2 & 1\end{bmatrix}$ to each **row**.

c. Create final result image $G$ according to the above formulation. Plot your resulting image in grayscale.

```python
#load test-image.jpg
image = imread('test-image.jpg')
n_rows,n_cols = image.shape

#Code for part a.
img_gx = np.zeros(image.shape)
for i in range(0, n_rows):
    img_gx[i,:] = signal.convolve(image[i,:], np.array([1,0,-1]), 'same')
for j in range(0, n_cols):
    img_gx[:,j] = signal.convolve(img_gx[:,j], np.array([1,2,1]), 'same')
plt.figure(figsize=(12,4))
plt.imshow(img_gx,'gray')
plt.title('Image Gx')

#Code for part b:
img_gy = np.zeros(image.shape)
for j in range(0, n_cols):
    img_gy[:,j] = signal.convolve(image[:,j], np.array([1,0,-1]), 'same')
for i in range(0, n_rows):
    img_gy[i,:] = signal.convolve(img_gy[i,:], np.array([1,2,1]), 'same')
plt.figure(figsize=(12,4))
plt.imshow(img_gy,'gray')
plt.title('Image Gy')

#Code for part c:
img_g = np.zeros(image.shape)
for i in range(0, n_rows):
    for j in range(0, n_cols):
        img_g[i,j] = np.sqrt(img_gx[i,j]**2 + img_gy[i,j]**2)
plt.figure(figsize=(12,4))
plt.imshow(img_g,'gray')
plt.title('Final Result Image G')
```

a. $y_1[n] = \frac{1}{4}x[n]-\frac{1}{16}x[n-1]+x[n-2]-

c. $y_3[n] = x[n] - \frac{1}{2}x[n-2] - y_3[n-3], \quad 0\leq n < 20$

```python
#Code for exercise 2:
b_a = [1/4, -1/16, 1, -1/16, 1/4]
a_a = [1, 0, 0, 0, 0]
z_a, p_a, _ = signal.tf2zpk(b_a,a_a)
n_a, y_a = signal.dimpulse((b_a,a_a,1), n=8)
h_a = y_a[0]

pz_plot(z_a, p_a, 'Part (a) Pole-Zero Plot')
plt.figure()
plt.stem(n_a, h_a)
plt.xlabel('n')
plt.ylabel('Impulse Response h[n]')
plt.title('System (a) Impulse Response')

b_b = [1, 0, 1/4]
a_b = [1, 2, -8]
z_b, p_b, _ = signal.tf2zpk(b_b,a_b)
n_b, y_b = signal.dimpulse((b_b,a_b,1), n=8)
h_b = y_b[0]

pz_plot(z_b, p_b, 'Part (b) Pole-Zero Plot')
plt.figure()
plt.stem(n_b, h_b)
plt.xlabel('n')
plt.ylabel('Impulse Response h[n]')
plt.title('System (b) Impulse Response')

b_c = [1, 0, -1/2, 0]
a_c = [1, 0, 0, 1]
z_c, p_c, _ = signal.tf2zpk(b_c,a_c)
n_c, y_c = signal.dimpulse((b_c,a_c,1), n=8)
h_c = y_c[0]

pz_plot(z_c, p_c, 'Part (c) Pole-Zero Plot')
plt.figure()
plt.stem(n_c, h_c)
plt.xlabel('n')
plt.ylabel('Impulse Response h[n]')
plt.title('System (c) Impulse Response')
```

Comments for 2.a:
BIBO stable because the ROC is given by |z| > 0 which includes the unit circle |z| = 1

Comments for 2.b:
Not BIBO stable because there is a pole at |z| = 2 and so the ROC is given by |z| > 2 which does not include the unit circle |z| = 1

Comments for 2.c:
Marginally stable because of there are poles (including a double pole) lying on the unit circle and so the ROC is given by |z| > 1.

## Exercise 3: Windows and Spectral Resolution

**Note: please specify 512 points for each FFT you take in this problem!**

a. Plot the magnitude spectrum (not dB scale) of the following signal using $\textrm{np.fft.rfft()}$.

$$
x_1[n] = 0.5\sin\left(0.5\pi n\right) + 0.02\sin\left(0.6\pi n\right), \quad 0\leq n < 80
$$

What is happening to the second sinusoid with the smaller magnitude in the frequency domain? Is it easy to locate this frequency peak?

b. Fill in the function $\textrm{modify()}$ below which applies a Hamming window to an input signal. Apply this function to $x_1[n]$ and plot the resulting magnitude spectrum. Is it easier to locate the smaller frequency now?

c. Now let's try resolving two close, but equally large frequency peaks in the below signal $x_2[n]$.

$$
x_2[n] = 0.5\sin\left(0.6\pi n\right) + 0.5\sin\left(0.618\pi n\right), \quad 0\leq n < 80
$$

Plot the magnitude spectrum of $x_2[n]$ (not dB scale) before and after applying the $\textrm{modify()}$ function. Is it easier to differentiate between the two peaks after applying the Hamming window? Why or why not? **Hint: think about the tradeoff between rectangular and Hamming windows!**


```python
#Code for part 3.a:
x1 = 0.50*np.sin(0.50*np.pi*np.arange(80)) + 0.02*np.sin(0.60*np.pi*np.arange(80))
fft_x1 = np.fft.rfft(x1, 512)
w_x1 = np.linspace(0,np.pi,len(fft_x1))

plt.figure(figsize=(12,4))
plt.title('FFT Magnitude Spectrum of x1')
plt.plot(w_x1, np.abs(fft_x1))
plt.xlabel('Frequency (rad/s)')
plt.ylabel('Magnitude Response |X1(w)|')

#Code for part 3.b:
def hamming(x):
    #apply a hamming window to the signal
    hamming_window = signal.windows.hamming(80)
    modified = hamming_window * x
    return modified

fft_x1_hamming = np.fft.rfft(hamming(x1), 512)
w_x1_hamming = np.linspace(0,np.pi,len(fft_x1_hamming))

plt.figure(figsize=(12,4))
plt.title('FFT Magnitude Spectrum of x1 with Hamming Window')
plt.plot(w_x1_hamming, np.abs(fft_x1_hamming))
plt.xlabel('Frequency (rad/s)')
plt.ylabel('Magnitude Response |X1(w)|')

#Code for part 3.c:
x2 = 0.50*np.sin(0.60*np.pi*np.arange(80)) + 0.50*np.sin(0.618*np.pi*np.arange(80))
fft_x2 = np.fft.rfft(x2, 512)
w_x2 = np.linspace(0,np.pi,len(fft_x2))

plt.figure(figsize=(12,4))
plt.title('FFT Magnitude Spectrum of x2 (before applying Hamming window)')
plt.plot(w_x2, np.abs(fft_x2))
plt.xlabel('Frequency (rad/s)')
plt.ylabel('Magnitude Response |X2(w)|')

fft_x2_hamming = np.fft.rfft(hamming(x2), 512)
w_x2_hamming = np.linspace(0,np.pi,len(fft_x2_hamming))

plt.figure(figsize=(12,4))
plt.title('FFT Magnitude Spectrum of x2 with Hamming Window')
plt.plot(w_x2_hamming, np.abs(fft_x2_hamming))
plt.xlabel('Frequency (rad/s)')
plt.ylabel('Magnitude Response |X2(w)|')
```

Comments for part 3.a:
No, it is not easy to locate the peak of the second sinusoid with the smaller magnitude since there are several other smaller peaks nearby. It is difficult to distinguish between these smaller peaks and identify the peak of the second sinusoid.

Comments for part 3.b:
Yes, it is easy to now locate the peak of the second sinusoid with the smaller magnitude due to the strong side lobe attenuation of the Hamming window. The smaller second peak is to the right of the peak of the first sinusoid. 

Comments for part 3.c:
No, it is not easy to distinguish the two peaks after applying the Hamming window. Compared to the Hamming window, the rectangular window has a narrower transition band and weaker stopband attenuation. As a result, it is possible to distinguish the two peaks due to the narrow transition band. While Hamming window has stronger stopband attenuation, its transition band is wider, making it difficult to distinguish the two peaks.

Comments for 5.c:
No, the simple filtering methods will not be able to perfectly separate the noise from the filtered audio.
The noise spectrum covering the entire frequency range and having higher magnitudes than the audio signal near the edges of the filtered audio makes complete signal recovery impossible. Although a low-pass filter can eliminate the high-frequency noise components, it may introduce noise of its own, and filtering components below a threshold can reduce very low frequency noise. Nevertheless, portions of the signal that are below the noise level will be lost.
